```python
climate_data = [
    {"city": "City A", "temperature": 25, "carbon_footprint": 500},
    {"city": "City B", "temperature": 30, "carbon_footprint": 350},
    {"city": "City C", "temperature": 22, "carbon_footprint": 600},
    {"city": "City D", "temperature": 15, "carbon_footprint": 200},
    {"city": "City E", "temperature": 28, "carbon_footprint": 450},
]
```

```python
high_temp_threshold=25

high_temp_cities = [city["city"] for city in climate_data if city["temperature"] > high_temp_threshold]

print(high_temp_cities)
for city in climate_data:
    if city["temperature"] > high_temp_threshold:
        print(city["city"])
```

```python
total_carbon_footprint = 0
for city in climate_data:
    total_carbon_footprint += city["carbon_footprint"]

average_carbon_footprint = total_carbon_footprint / len(climate_data)

print(f"The average carbon footprint across all cities is: {average_carbon_footprint}")
```

```python
highest_footprint_city = max(climate_data, key=lambda x: x["carbon_footprint"])
print("Highest carbon footprint city:", highest_footprint_city['city'])
```

```python
year = int(input("Enter a year: "))
def leap(year):
    return (lambda y: y % 4 == 0 and (y % 100 != 0 or y % 400 == 0))(year)
if leap(year):
    print(f"{year} is a leap year.")
else:
    print(f"{year} is not a leap year.")
```

```python
import pandas as pd
import numpy as np
data = {
    "Energy Source": ["Solar", "Wind", "Hydropower", "Geothermal", "Biomass", "Nuclear"],
    "Energy Consumption (MWh)": [1200, np.nan, 2900, np.nan, 2500, 3200],
    "Cost (Million $)": [200, 400, np.nan, 150,250,np.nan]
}
energy_df = pd.DataFrame(data)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
energy_df[["Energy Consumption (MWh)","Cost (Million $)"]] = scaler.fit_transform(
    energy_df[["Energy Consumption (MWh)","Cost (Million $)"]]
)
print("\nData After Normalization (XZ- score Scaling): ")
print(energy_df)
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder

data = {
    "Energy Source": ["Solar", "Wind", "Hydropower", "Geothermal", "Biomass", "Nuclear"],
    "Energy Consumption (MWh)": [1200, np.nan, 2900, np.nan, 2500, 3200],
    "Cost (Million $)": [200, 400, np.nan, 150,250,np.nan]
}
energy_df = pd.DataFrame(data)

# Create a OneHotEncoder object
encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')

# Fit and transform the 'Energy Source' column
encoded_data = encoder.fit_transform(energy_df[['Energy Source']])

# Create a new DataFrame with the one-hot encoded features
encoded_df = pd.DataFrame(encoded_data, columns=encoder.get_feature_names_out(['Energy Source']))

# Concatenate the one-hot encoded features with the original DataFrame
energy_df = pd.concat([energy_df, encoded_df], axis=1)

# Display the updated DataFrame
energy_df
```
