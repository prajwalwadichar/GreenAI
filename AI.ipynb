```python
#use lambda function to check given year is leap year or not

is_leap = lambda year: (year % 4 == 0 and year % 100 != 0) or year % 400 == 0

print(is_leap(2000))
print(is_leap(1900))
print(is_leap(2024))
```

```python
#to check weather a strring is palindeome or not using numpy
import numpy as np

def is_palindrome(s):
    s = s.replace(" ", "").lower()

    arr = np.array(list(s))

    return np.array_equal(arr, arr[::-1])

string = "madam"
if is_palindrome(string):
    print(f"'{string}' is a palindrome.")
else:
    print(f"'{string}' is not a palindrome.")
```

```python
# Create an array for energy consumption (in kWh)
energy_consumption = np.array([1200, 3400, 2900, 1800, 2500])

# Calculate the mean
mean_consumption = np.mean(energy_consumption)
print("Mean energy consumption (kWh):", mean_consumption)

# Calculate the median
median_consumption = np.median(energy_consumption)
print("Median energy consumption (kWh):", median_consumption)

# Calculate the standard deviation
std_dev_consumption = np.std(energy_consumption)
print("Standard deviation of energy consumption (kWh):", std_dev_consumption)

# Calculate the variance
variance_consumption = np.var(energy_consumption)
print("Variance of energy consumption (kWh):", variance_consumption)
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

#creating a dataframe for pair plot
data={
    "Solar": [1200, 1500, 1300],
    "Wind": [3400, 3600,3200],
    "Hydropower": [2900, 3100, 2800],
    "Biomass": [2500, 2700, 2400]
}
df =pd.DataFrame(data)
#a Ceate a pair plot
sns.pairplot(df)
plt.show()
# create a hit map to show co relation matrix
# Ensure all columns are numeric before calculating correlation
for col in df.columns:
    if not pd.api.types.is_numeric_dtype(df[col]):
        print(f"Column {col} is not numeric, dropping")
        df = df.drop(col, axis=1) # Drop non-numeric column

correlation_matrix=df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", linewidth=0.5)
plt.title('Correlation Between energy sources')
plt.show()
```

```python
#create a box plot
sns.boxplot(data=df)

#add title
plt.title("Distribution of Energy Consumption by Source")

#show the plot
plt.show()
```

```python
#create a violin plot
sns.violinplot(data=df)
#add title
plt.title('Violin plot of Energy Consumption Distributions')
plt.show()
```
